```python
import cv2
import os
import numpy as np
```

### Extracting Images from the dataset:

```python
images_list=[]
classification_label=[]
class_name=[]
```

#### defining parent folder address:

```python
base_dir="plantvillage dataset/color"
iter=0
for i in os.listdir(base_dir):
    # print(i)
    class_name.append(i)
    classification_label.append(iter)
    iter=iter+1
```

#### List of classes and its corresponding numerical labels:

```python
for i,j in zip(classification_label,class_name):
    print(f"{i} - {j}")
```

```python
directories_list=[]
for i in class_name:
    directory=base_dir+'/'+i
    directories_list.append(directory)
```

```python
for i in directories_list:
    print(i)
```

```python
for i in os.listdir(directories_list[0]):
    print(i)
```

```python
img_size=256
```

```python
iter=0
for p in directories_list:
    count=0
    for q in os.listdir(p):
        img_path=p+'/'+q
        img=cv2.imread(img_path)
        if img is not None:
            img=cv2.resize(img,(img_size,img_size)) #rescaling the size of images
            img=(img/255.0).astype('float32') #normalizing the pixel values of images
            images_list.append(img)
            classification_label.append(iter)
            print(f"class {iter} - image {count} updated")
            count=count+1
    iter=iter+1

print("checkpoint")
```
